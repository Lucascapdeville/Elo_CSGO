```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import re
import calendar
```

# Majors

```python
match_page = []
event_types = 'Majors'
for i in range(12,22):
    for offset in range(0,951,50):
        url_matches = f'https://www.hltv.org/stats/matches?startDate=20{i}-01-01&endDate=20{i}-12-31&matchType={event_types}&offset={offset}'
        print(url_matches)
        html_matches = requests.get(url_matches)
        print(html_matches)
        tabela = pd.read_html(html_matches.content)[0]
        if pd.read_html(html_matches.content)[0].shape[0] == 0:
            print(offset+1)
            break
        else:
            tabela['event_type'] = event_types
            tabela['event_tier'] = 'S'
            match_page.append(tabela)
    try:
        matches = pd.concat(match_page)
        matches.reset_index(drop=True,inplace= True)
        matches.to_csv(f'data/matches_{event_types}_{i}.csv',sep = ';')
    except:
        pass
matches = pd.concat(match_page)
matches.reset_index(drop=True,inplace= True)
matches.to_csv(f'data/matches_{event_types}.csv',sep = ';')
```
